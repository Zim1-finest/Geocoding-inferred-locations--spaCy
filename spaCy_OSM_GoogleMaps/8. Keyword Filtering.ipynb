```python
df = pd.read_csv('data//analysis_data//all_locations_extracted.csv')
```

```python
# defining keywords for historial mentions 

past_keywords = ['arrived from ', 'had gone to ', 'coming from ',
                 'I was in ', 'we were in ', 'I was at ', ' was at ', ' was in ', ' were at ', ' went to ',
                 "landed from ", 'passed through', ' had visited ', ' had gone to', "flew from ", 'flew in from',
                 ' back from ',  ' miss being in ', ' was leaving in ', ' was staying at ', ' was leaving at ',
                 ' was staying in ', ' im from ', ' am from ', ' are from ', 'moved from',
                 ' originally from ', 'grew up in ' ]
past_searched = '|'.join(past_keywords) # for searching keywords within sentence structures

# Filtering the data to return only data with the specific keyword
# case = False makes the search case insensitive 
# na = false means we dont return errors when there are unexpected types in series 

df_past = df[df["en_text"].str.contains(past_searched, case = False, na = False)]
df_past.shape
```

```python
past_searched
```

```python
df_past[['en_text','source']].tail(10)
```

```python
# We add space before and after 'to' to make the text standalone

future_keywords = [' driving to', 'Taking the train to'," taking the car to"," taking the bus to",
                   ' heading to', ' headed for', "leave for", " leaving for ", ' will go to ', ' travel to', 'trip to',
                   ' travelling to ', ' moving to', ' relocating to ', ' flying to ',' will be going to', ' will be at ',
                   ' will be in ',' next stop ', ' move to ', ' transferring to ', ' off to ', 'departing for ', 
                   'on my way to', 'on our way to']
future_searched = '|'.join(future_keywords) # for searching keywords within sentence structures

# filtering data to return only future mentions 

df_future = df[df["en_text"].str.contains(future_searched, case = False, na = False)]
df_future.shape
```

```python
future_searched
```

```python
df_future[['en_text','source']].head(10)
```

```python
# df-final includes all present tweets and unclassified tweets. Removes tweets classified as past or future 
df_final = df[~df["en_text"].str.contains(future_searched, case = False, na = False)& ~df["en_text"].str.contains(past_searched, case = False, na = False)]
df_final.shape
```

```python
df_final.to_csv('data//analysis_data//keyword_filtered_posts.csv')
```

# Temporal Information

```python
# replace all NaN with empty string
df_final.TE = df_final.TE.replace(np.nan, '', regex=True)
```

```python
df_TE = df_final[(df_final['TE'] != '')]
df_TE.shape
```

```python
TE_discard = ['tomorrow', 'yesterday', 'November', 'December','September', 'the other day',
              'Festival2017','Saturday, 2/15', 'March 28, 2009', 'last fall',
              'two more days', '1998-2019', '2019, an amazing year', 'just the previous day',
              'June 1981 ','April 1995 ', 'years back', 'oct ', 'Xmas ',
              'August 21, 2008 - October 31, 2010', 'From 1920 to 1961', 'Saturday, August 18th', 'April 27, 2013',
               '12/26/2019 ','1980 - 1992 ', '03/06/2020', 'March 1993', 'Feb. 9', 'early March',
              'weeks ago', 'Saturday, 1992','Christmas', '06-14-2019', '3/25/20', 'ago', 'past', 'from now',
              'last', 'next', 'this', 'back', 'coming', '7/9/2010 ','tmrw', 'Dec 1, 2015 ','Nov', 'Straturday, 2007',
              'July 9-, 2009, 2015', 'January 21, 2019', 'TBT', 'early', 'October'
              
             ]

discard_searched = '|'.join(TE_discard) # for searching keywords within sentence structures

# filtering data to return only current mentions 

df_discard = df_final[df_final["TE"].str.contains(discard_searched, case = False, na = False)]
df_discard.shape
```

```python
TE_keep = ['today','this year','this week','the weekend', 'these days']

keep_searched = '|'.join(TE_keep) # for searching keywords within sentence structures

# filtering data to return only current mentions 

df_keep = df_final[df_final["TE"].str.contains(keep_searched, case = False, na = False)]
df_keep.shape
```

```python
# df-final includes all present tweets and unclassified tweets. Removes tweets classified as past or future 
df_final2 = df_final[~df_final["TE"].str.contains(discard_searched, case = False, na = False) | df["TE"].str.contains(keep_searched, case = False, na = False)]
df_final2.shape
```

```python
df_final2.to_csv('data//analysis_data//keyword+temp_filtered_posts.csv')
```
